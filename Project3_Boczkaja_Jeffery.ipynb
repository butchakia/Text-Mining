```python
import PyPDF2
import re
import pdfreader
```

### Problem 1: Extract all texts from the pdf file.

```python
pdf = 'tax-efficient-withdrawal-strategies.pdf'
pdfFile = open(pdf, 'rb')
```

```python
pdfFileReader = PyPDF2.PdfReader(pdf)
```

```python
pageCount = len(pdfFileReader.pages)
print(f' There are {pageCount} pages in the file :{pdf}')
```

```python
output = []
for i in range(pageCount):
    pdfPage = pdfFileReader.pages[i]
    output.append(pdfPage.extract_text())
```

```python
alltexts = ' '.join(output)
```

### Problem 2: Extract all the tokens and lemmas from the texts.

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(alltexts)
for token in doc:
    print(f'Token text = {token.text}; Is the token lowercase? {token.is_lower}; Does the token consit of digits? {token.is_digit} ')
```

```python
for token in doc:
    print(f'Token text = {token.text}; The lemma ={token.lemma_} ')
```

## Problem 3: Remove all the default stop words in SpaCy from the texts.

```python
nlp = spacy.load('en_core_web_sm')
spacy_stopwords = nlp.Defaults.stop_words
doc = nlp(alltexts)
tokens_without_stopword= [token for token in doc if not token.text in spacy_stopwords]

print(tokens_without_stopword[:100])
```

### Problem 4: Customize the stop words in SpaCy.

```python
print(f'There are {len(nlp.Defaults.stop_words)} stop words in spaCy')
customized_stop_words = ['tax', 'account']

for token in customized_stop_words:
    nlp.Defaults.stop_words.add(token)

for token in customized_stop_words:
    nlp.vocab[token].is_stop = True
    
print(f'There are {len(nlp.Defaults.stop_words)} stop words in spaCy')
```

```python
print(f'There are {len(nlp.Defaults.stop_words)} stop words in spaCy')
remove_stop_words = ['full']

for token in remove_stop_words:
    nlp.Defaults.stop_words.remove(token)

for token in remove_stop_words:
    nlp.vocab[token].is_stop = False

print(f'There are {len(nlp.Defaults.stop_words)} stop words in spaCy')
```

```python
print(f'There are {len(nlp.Defaults.stop_words)} stop words in spaCy')
remove_stop_words = ['tax', 'account']

for token in remove_stop_words:
    nlp.Defaults.stop_words.remove(token)

for token in remove_stop_words:
    nlp.vocab[token].is_stop = False

print(f'There are {len(nlp.Defaults.stop_words)} stop words in spaCy')
```

### Problem 5: Perform the part of speech tagging for the texts.

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(alltexts)

for token in doc[:20]:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)

# To see all pos tagging run:
# for token in doc:
#    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
#            token.shape_, token.is_alpha, token.is_stop)
```

### Another way to perform part of speech tagging for the texts.

```python
pos_output = [(token.text, token.pos_) for token in doc]
for token, pos in pos_output:
    print(f'Token: {token}, POS: {pos}')
```

###  Problem 7: Perform the named entities recognition for the texts.

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```
